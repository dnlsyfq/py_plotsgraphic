```python
import csv
from itertools import groupby
```

```python
input_file = 'iris.csv'
```

```python
with open(input_file,'r') as iris_data:
    irises = list(csv.reader(iris_data))
```

```python
### scatter
colors = {
    "setosa":"#2B5B84",
    "versicolor":"g",
    "virginica":"purple"
}

irises[:3]
```

```python
sepal_length = []
sepal_width = []
petal_length = []
species = []

for i in irises[1:]:
    sepal_length.append(float(i[0]))
    sepal_width.append(float(i[1]))
    petal_length.append(float(i[2]))
    species.append(i[4])
```

```python
petal_lengths = []
species = []

petal_dict = {}

for i in irises[1:]:
    if i[4] not in petal_dict:
        petal_dict[i[4]] = []
    else:
        petal_dict[i[4]].append(float(i[3]))
```

```python
petal_lengths = []
for species, group in groupby(irises[1:], lambda i:i[4]):
    petal_lengths.append([float(petal[2]) for petal in group])

plt.boxplot(petal_lengths)
plt.axis([0,4,0,10])
plt.xticks([1,2,3],['Setosa','Versicolor','Virginica'])
plt.title("Fisher's Iris Data Set, Petal",fontsize=12,loc='left')
plt.xlabel("Iris Variety",fontsize=10)
plt.ylabel("Petal length (cm)",fontsize=10)
plt.show()
```

```python
for species, group in groupby(irises[1:], lambda i:i[4]):
    categorized_irises = list(group)
    sepal_lengths = [float(iris[0]) for iris in categorized_irises]
    sepal_widths = [float(iris[1]) for iris in categorized_irises]
    plt.scatter(sepal_lengths,sepal_widths,s=10,c=colors[species],label=species)

plt.title("Fisher's Iris Data Set",fontsize=12)
plt.xlabel("sepal length",fontsize=10)
plt.ylabel("sepal width",fontsize=10)
plt.legend(loc="upper right")
plt.show()
```

```python
import csv
```

```python
input_file = 'old_faithful.csv'
plt.figure(figsize=(7.5,4.25))
plt.style.use('classic')

with open(input_file,'r') as old_faithful_data:
    eruptions = list(csv.reader(old_faithful_data))
```

```python
eruptions[:3]
```

```python
eruptions_times = []
waiting_times = []
for i in eruptions[1:]:
    eruptions_times.append(float(i[0]))
    waiting_times.append(float(i[1]))
```

```python
plt.subplot(2,2,1)
plt.boxplot(eruptions_times)
plt.title('Old Faithful Eruptions')
plt.xticks([1],['Eruptions'])
plt.xlabel('Length of Eruption (mins)')
```

```python
plt.subplot(2,2,2)
num_bins = 8
plt.hist(eruptions_times,num_bins)
plt.title('Old Faithful Eruptions')
plt.xlabel('Length of Eruption (mins)')
```

```python
plt.subplot(2,2,3)
plt.boxplot(waiting_times)
plt.title('Old Faithful Waiting')
plt.xticks([1],['Waiting'])
plt.xlabel('Length of Waiting (mins)')
```

```python
plt.subplot(2,2,4)
plt.scatter(eruptions_times,waiting_times)
plt.title('Old Faithful Eruptions')
plt.xlabel('Length of Eruption (mins)')
plt.ylabel('Time Between Eruptions')
```

```python
plt.tight_layout()
plt.show()
```
